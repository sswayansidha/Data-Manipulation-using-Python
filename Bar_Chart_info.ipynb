```python
import pandas as pd
```

```python
data = {'ORDERNUMBER': ['A_10107', 'A_10121', 'A_10134', 'A_10145', 'A_10159','A_10168','A_10180','A_10188'],
        'QUANTITYORDERED': [30, 34, 41, 45, 49,36,29,48]
        }

df1 = pd.DataFrame(data)
```

```python
df1
```

```python
import matplotlib.pyplot as plt
```

**Approch 1**

```python
order_no = list(df1.ORDERNUMBER)
```

```python
qty = list(df1.QUANTITYORDERED)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.bar(order_no,qty)
plt.show()
```

**Approch 2**:  Customization

```python
fig = plt.figure(figsize=(10,5))
plt.bar(order_no,qty,color='orange',width = 0.8)
plt.xlabel("Order ID")
plt.ylabel("Order Quantity")
plt.title("Quantity vs Orders")
plt.show()
```

**Label the Bar**

```python
fig = plt.figure(figsize=(10,5))
plt.bar(order_no,qty,color='pink',width = 0.8)
plt.xlabel("Order ID")
plt.ylabel("Order Quantity")
plt.title("Quantity vs Orders")
#-------------------------------------------------------------------------------
for index,data in enumerate(qty):
  plt.text(x=index, y = data+1 , s=f"{data}",fontdict=dict(fontsize=20))
#-------------------------------------------------------------------------------
plt.show()
```

**Define the axis length**



```python
fig = plt.figure(figsize=(10,5))
plt.bar(order_no,qty,color='green',width = 0.8)
plt.xlabel("Order ID")
plt.ylabel("Order Quantity")
plt.title("Quantity vs Orders")

for index,data in enumerate(qty):
  plt.text(x=index, y = data+1 , s=f"{data}",fontdict=dict(fontsize=20))
#-------------------------------------------------------------------------------
max_y_lim = max(qty)+10
min_y_lim = 0
plt.ylim(min_y_lim,max_y_lim)
#-------------------------------------------------------------------------------
plt.show()
```

**Bring Lebels to the centre of the bar**

```python
fig = plt.figure(figsize=(10,5))
plt.bar(order_no,qty,color='brown',width = 0.8)
plt.xlabel("Order ID")
plt.ylabel("Order Quantity")
plt.title("Quantity vs Orders")
#----------------------------------------------------------------------------------------------------------------
for index,data in enumerate(qty):
  plt.text(x=index, y = data+1 , s=f"{data}",fontdict=dict(fontsize=20),horizontalalignment='center')
#----------------------------------------------------------------------------------------------------------------
max_y_lim = max(qty)+10
min_y_lim = 0
plt.ylim(min_y_lim,max_y_lim)

plt.show()
```

**Rotate the Labels vertically**

```python
fig = plt.figure(figsize=(10,5))
plt.bar(order_no,qty,color='gray',width = 0.8)
plt.xlabel("Order ID")
plt.ylabel("Order Quantity")
plt.title("Quantity vs Orders")
#---------------------------------------------------------------------------------------------------------------------------
for index,data in enumerate(qty):
  plt.text(x=index, y = data+1 , s=f"{data}",fontdict=dict(fontsize=20),horizontalalignment='center',rotation='vertical')
#----------------------------------------------------------------------------------------------------------------------------
max_y_lim = max(qty)+10
min_y_lim = 0
plt.ylim(min_y_lim,max_y_lim)

plt.show()
```

**Removing Ticks**

```python
fig = plt.figure(figsize=(10,5))
plt.bar(order_no,qty,color='navy',width = 0.8)
plt.xlabel("Order ID")
plt.ylabel("Order Quantity")
plt.title("Quantity vs Orders")

for index,data in enumerate(qty):
  plt.text(x=index, y = data+1 , s=f"{data}",fontdict=dict(fontsize=20),horizontalalignment='center')

max_y_lim = max(qty)+10
min_y_lim = 0
plt.ylim(min_y_lim,max_y_lim)
#---------------------------------------------------------------------------------------
plt.tick_params(bottom=False) # Use "left" instead of bottom for y axis ticks
#---------------------------------------------------------------------------------------
plt.show()
```

**Custom Tick**

```python
fig = plt.figure(figsize=(10,5))
plt.bar(order_no,qty,color='yellow',width = 0.8)
plt.xlabel("Order ID")
plt.ylabel("Order Quantity")
plt.title("Quantity vs Orders")

for index,data in enumerate(qty):
  plt.text(x=index, y = data+1 , s=f"{data}",fontdict=dict(fontsize=20),horizontalalignment='center')

max_y_lim = max(qty)+10
min_y_lim = 0
plt.ylim(min_y_lim,max_y_lim)
#---------------------------------------------------------------------------------------
plt.tick_params(axis='y', colors='red', direction='out', length=13, width=3)
plt.tick_params(bottom=False)
#---------------------------------------------------------------------------------------
plt.show()
```

**Sort dataframe with respect to a column**

```python
df2 = df1.sort_values('QUANTITYORDERED',ascending=False)
```

```python
df2
```

**Sort Bars and add border to bars**

```python
fig = plt.figure(figsize=(10,5))
#---------------------------------------------------------------------------------------
order_no2 = list(df2.ORDERNUMBER)
qty2 = list(df2.QUANTITYORDERED)
#---------------------------------------------------------------------------------------
plt.xlabel("Order ID")
plt.ylabel("Order Quantity")
plt.title("Quantity vs Orders")

max_y_lim = max(qty2)+10
min_y_lim = 0
plt.ylim(min_y_lim,max_y_lim)
plt.tick_params(bottom=False) # Use "left" instead of bottom for y axis ticks
#--------------------------------------------------------------------------------------------
plt.bar(order_no2,qty2,color='#F0F8FF',width = 0.8,edgecolor = "red")
#------------------------------------------------------------------------------------------
for index,data in enumerate(qty2):
  plt.text(x=index, y = data+1 , s=f"{data}",fontdict=dict(fontsize=20),horizontalalignment='center')

plt.show()
```

**Bring the lebels to bottom of the bar**

```python
fig = plt.figure(figsize=(10,5))

order_no2 = list(df2.ORDERNUMBER)
qty2 = list(df2.QUANTITYORDERED)

plt.xlabel("Order ID")
plt.ylabel("Order Quantity")
plt.title("Quantity vs Orders")

max_y_lim = max(qty2)+10
min_y_lim = 0
plt.ylim(min_y_lim,max_y_lim)
plt.tick_params(bottom=False) 

plt.bar(order_no2,qty2,color='#01A9DB',width = 0.8,edgecolor = "red")
#------------------------------------------------------------------------------------------

for index, data in enumerate(qty2):
    plt.text(index, data/qty2[index]+(0), qty2[index], fontsize=18,horizontalalignment='center')
#--------------------------------------------------------------------------------------------
plt.show()
```

**Draw Grid - Transparent**

```python
fig = plt.figure(figsize=(10,5))

order_no2 = list(df2.ORDERNUMBER)
qty2 = list(df2.QUANTITYORDERED)

plt.xlabel("Order ID")
plt.ylabel("Order Quantity")
plt.title("Quantity vs Orders")

max_y_lim = max(qty2)+10
min_y_lim = 0
plt.ylim(min_y_lim,max_y_lim)
plt.tick_params(bottom=False) 

#--------------------------------------------------------------------------------------------
plt.grid(color = "grey", linewidth = "1.4", linestyle = "-.",alpha = 0.2)
#------------------------------------------------------------------------------------------
plt.bar(order_no2,qty2,color='#01A9DB',width = 0.8,edgecolor = "red",alpha=1)

for index, data in enumerate(qty2):
    plt.text(index, data/qty2[index]+(0), qty2[index], fontsize=18,horizontalalignment='center')

plt.show()
```
